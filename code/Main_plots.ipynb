```python
Rate_T1=loadtxt('cum_rate_per_episode_T1.csv', delimiter=',')
Rate_T2=loadtxt('cum_rate_per_episode_T2.csv', delimiter=',')
Rate_OMA=loadtxt('cum_rate_per_episode_OMA.csv', delimiter=',')
Rate_SD=loadtxt('cum_rate_per_episode_SD.csv', delimiter=',')
Rate_R1=loadtxt('cum_rate_per_episode_R1.csv', delimiter=',')

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
bar_labels = ['NOMA-DQN (10dB)', 'NOMA-DQN (15dB)', 'Shortest Path', 'OMA','Random selection']
bar_colors = ['red', 'blue', 'green', 'purple','cyan']

Algorithms = ['NOMA-DQN (10dB)', 'NOMA-DQN (15dB)', 'Shortest Path', 'OMA','Random selection']
Average_rate = [np.mean(Rate_T1), np.mean(Rate_T2), np.mean(Rate_SD), np.mean(Rate_OMA),np.mean(Rate_R1)]
ax.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.3)
ax.bar(Algorithms, Average_rate,width = 0.3,label=bar_labels, color=bar_colors)


ax.set_xticks(bar_labels,labels=bar_labels, fontsize=6)
ax.set_ylabel('Average Data Rate (Mb/s)')
ax.set_xlabel('Algorithms')
ax.legend(bar_labels,fontsize="10")

ax.ticklabel_format(axis='y', style='sci', scilimits=(2,2))
plt.show()

ax.figure.savefig(r'C:\Users\student\OneDrive - Carleton University (1)\Erricson\Third paper\Simulation_DRL\Plots\Avg_rate_bar.pdf')
```
